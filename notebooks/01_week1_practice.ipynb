Bài 1: Thống kê danh sách

```python
values = [12, 7, 9, 12, 15, 7, 21, 9, 12]

frep = {}

for x in values:
    if x in frep:
        frep[x] += 1
    else:
        frep[x] = 1
print(frep)
```

Bài 2: Lọc và biến đổi dữ liệu

```python
prices_usd = [10.5, 25.0, 5.0, 42.0, 8.5, 15.0]

prices_vnd = []

for p in prices_usd:
    if p > 10.0:
        p_vnd = p * 25000
        prices_vnd.append(p_vnd)
print(prices_vnd)
```

```python
prices_usd = [10.5, 25.0, 5.0, 42.0, 8.5, 15.0]

prices_vnd = [p * 25000 for p in prices_usd if p > 10.0]

print(prices_vnd)
```

Bài 3: Làm sạch dữ liệu

```python
raw_name = [" Alice", "BOB", "", " alice", None, "Charlie"]

clean_names = []

for name in raw_name:
    if name is not None:
        cleaned = name.strip().lower()
        if cleaned != "" and cleaned not in clean_names:
            clean_names.append(cleaned)

print(clean_names)
```

Bài 4: Data hợp nhất

```python
names = ["An", "Bình", "Chi"]
scores = [8.5, 6.0, 9.0]

result = []

for i in range(len(names)):
    item = {"name": names[i], "score": scores[i]}
    result.append(item)
    
print(result)
```

```python
names = ["An", "Bình", "Chi"]
scores = [8.5, 6.0, 9.0]

result = [{"name": name, "score": score} for name, score in zip(names, scores)]

print(result)
```
