# BÀI KIỂM TRA ĐẦU VÀO
*Chương trình thống kê và phân loại điểm số*

```python
scores = [7.5, 8.0, 5.5, 9.0, 6.5, 8.5, 4.0]

def process_scores(score_list):
    avg_score = sum(score_list) / len(score_list)
    max_score = max(score_list)
    min_score = min(score_list)
    count_high  = sum(1 for s in score_list if s >= 8.0)
    filtered_scores = [s for s in score_list if s >= 5.0]

    return {
        "mean": round(avg_score, 2),
        "max": max_score,
        "min": min_score,
        "high_scores_count": count_high,
        "passed_scores":filtered_scores,
    }

def classify_scores(score_list):
    results = []
    for s in score_list:
        if s >= 8.0:
            grade = "Excellent"
        elif s >= 6.5:
            grade = "Good"
        elif s >= 5.0:
            grade = "Average"
        else:
            grade = "Fail"
        results.append({"score": s, "grade": grade})
    return results

print("--- Statistical results ---")
print(process_scores(scores))

print("\n--- Classification results ---")
print(classify_scores(scores))
```
